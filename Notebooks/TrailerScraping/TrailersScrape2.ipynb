```python
from datetime import date
from datetime import datetime


def get_posterior_counts(comments, movie):
    movie_date_object = datetime.strptime(movie["release_start"], "%m/%d/%Y")

    posterior_comments_count = 0
    posterior_reply_count = 0

    for comment in comments:
        comment_updatedAt = comment["snippet"]["topLevelComment"]["snippet"]["updatedAt"].replace("T", " ").replace("Z", "")
        comment_date_object = datetime.strptime(comment_updatedAt, "%Y-%m-%d %H:%M:%S")

        comment_movie_days_diff = (comment_date_object-movie_date_object).days
        #print("Movie Release Date {}, Comment Date {}, days_diff {}".format(movie_date_object, comment_date_object, comment_movie_days_diff))
        if comment_movie_days_diff >= 0:
            posterior_comments_count+= 1

        for reply in comment["replies"]:
            reply_updatedAt = reply["snippet"]["updatedAt"].replace("T", " ").replace("Z", "")
            reply_date_object = datetime.strptime(reply_updatedAt, "%Y-%m-%d %H:%M:%S")
            reply_movie_days_diff = (reply_date_object-movie_date_object).days
            #print("Movie Release Date {}, Reply Date {}, days_diff {}".format(movie_date_object, reply_date_object, reply_movie_days_diff))
            if reply_movie_days_diff >= 0:
                posterior_reply_count+= 1

    return posterior_comments_count, posterior_reply_count
```

```python
import time
def get_comment_list(youtube, video_id, next_view_token):
    finish_succesfully = False
    retry_count = 0

    comment_list = {}
    comment_list["items"] = []

    while not finish_succesfully:
        try:
            if next_view_token == "":
                comment_list = youtube.commentThreads().list(part = 'snippet',
                                                     maxResults = 100,
                                                     videoId = video_id,
                                                     order = 'time').execute()
            else:
                comment_list = youtube.commentThreads().list(part = 'snippet',
                                                     maxResults = 100,
                                                     videoId = video_id,
                                                     order='time',
                                                     pageToken=next_view_token).execute()

            finish_succesfully = True
        except Exception as e:
            if "commentsDisabled" in str(e):
                break
            else:
                print(e)

            #print(e.args)
            #print(e["reason"])
            finish_succesfully = False
            retry_count += 1
            time.sleep(1)
            if retry_count == 1:
                print("Retry count 15 reached")
                break
    return comment_list

def get_replies_list(youtube, comment_id, next_view_token):
    finish_succesfully = False
    retry_count = 0
    replies_list = []
    while not finish_succesfully:
        try:
            if next_view_token == "":
                replies_list = youtube.comments().list(part='snippet',
                                                   maxResults=100,
                                                   parentId=comment_id).execute()
            else:
                replies_list = youtube.comments().list(part = 'snippet',
                                                       maxResults = 100,
                                                       parentId = comment_id,
                                                       pageToken = next_view_token).execute()

            finish_succesfully = True
        except Exception as e:
            print(e)
            #print(e.args)
            #print(e["reason"])
            finish_succesfully = False
            retry_count += 1
            time.sleep(1)
            if retry_count == 1:
                print("Retry count 1 reached")
                break
    return replies_list

def get_delete_comments_count(youtube, video_id, movie):
    total_posterior_comments_count = 0
    total_posterior_reply_count = 0

    iterations_in_zero_count = 0

    next_view_token = ""
    while True:
        iteration_comments_count = 0
        iteration_replies_count = 0
        iteration_comments = []
        if next_view_token == '':
            # get the initial response
            comment_list = get_comment_list(youtube, video_id, "")
        else:
            # get the next page response
            comment_list = get_comment_list(youtube, video_id, next_view_token)

        for comment in comment_list['items']:
            # get number of replies
            reply_count = comment['snippet']['totalReplyCount']

            all_replies = []

            # if replies greater than 0
            if reply_count > 0:
                # get first 100 replies
                replies_list = get_replies_list(youtube, comment['id'], "")

                for reply in replies_list['items']:
                    # add reply to list
                    all_replies.append(reply)
                    iteration_replies_count += 1

                # check for more replies
                while "nextPageToken" in replies_list:
                    token_reply = replies_list['nextPageToken']
                    # get next set of 100 replies
                    replies_list = get_replies_list(youtube, comment['id'], token_reply)

                    for reply in replies_list['items']:
                        # add reply to list
                        all_replies.append(reply)
                        iteration_replies_count += 1

            comment["replies"] = all_replies
            # add comment to list
            iteration_comments.append(comment)
            iteration_comments_count += 1

        posterior_comments_count, posterior_reply_count = get_posterior_counts(iteration_comments, movie)
        print("Iteration results: posterior_comments_count: {} from {}, posterior_reply_count: {} from {}".format(posterior_comments_count, iteration_comments_count, posterior_reply_count, iteration_replies_count))

        if posterior_comments_count == 0 and posterior_reply_count == 0:
            iterations_in_zero_count += 1
        else:
            total_posterior_comments_count += posterior_comments_count
            total_posterior_reply_count += posterior_reply_count

        if iterations_in_zero_count >= 5:
            break

        if "nextPageToken" in comment_list:
            next_view_token = comment_list['nextPageToken']
        else:
            break

    return total_posterior_comments_count, total_posterior_reply_count
```

```python
def days_diff(date1, date2):
    try:
        movie_moth, movie_day, movie_year = date2.split("/")
        result_moth, result_day, result_year = date1.split("/")
        movie_date = date(int(movie_year), int(movie_moth), int(movie_day))
        result_date = date(int(result_year), int(result_moth), int(result_day))
        days_diff_video_movie = (result_date - movie_date).days
    except Exception as e:
        days_diff_video_movie = 1
    return days_diff_video_movie

new_movies = []
for movie in movies:
    results = []
    for result in movie["results"]:
        result["days_diff"] = days_diff(result["release_date"], movie["release_start"])
        results.append(result)
    movie["results"] = results
    new_movies.append(movie)
movies = new_movies
```

```python
def get_comments(youtube, video_id, n_iterations):
    all_comments = []
    next_view_token = ""
    iteration = 0
    while True:
        if next_view_token == '':
            # get the initial response
            comment_list = youtube.commentThreads().list(part = 'snippet',
                                                     maxResults = 100,
                                                     videoId = video_id,
                                                     order = 'time').execute()
        else:
            # get the next page response
            comment_list = youtube.commentThreads().list(part = 'snippet',
                                                     maxResults = 100,
                                                     videoId = video_id,
                                                     order='time',
                                                     pageToken=next_view_token).execute()
        for comment in comment_list['items']:
            # get number of replies
            reply_count = comment['snippet']['totalReplyCount']

            all_replies = []

            # if replies greater than 0
            if reply_count > 0:
                # get first 100 replies
                replies_list = youtube.comments().list(part='snippet',
                                                   maxResults=100,
                                                   parentId=comment['id']).execute()
                for reply in replies_list['items']:
                    # add reply to list
                    all_replies.append(reply)

                # check for more replies
                while "nextPageToken" in replies_list:
                    token_reply = replies_list['nextPageToken']
                    # get next set of 100 replies
                    replies_list = youtube.comments().list(part = 'snippet',
                                                       maxResults = 100,
                                                       parentId = comment['id'],
                                                       pageToken = token_reply).execute()
                    for reply in replies_list['items']:
                        # add reply to list
                        all_replies.append(reply)

            comment["replies"] = all_replies
            # add comment to list
            all_comments.append(comment)
        iteration += 1

        if iteration == n_iterations:
            break

        if "nextPageToken" in comment_list:
            next_view_token = comment_list['nextPageToken']
        else:
            break

    return all_comments

comments = get_comments(youtube, video_id, 10)
```

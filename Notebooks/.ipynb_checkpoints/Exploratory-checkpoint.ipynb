# Reading Movies

```python
import ast
import csv
movies = []
csv_filename = 'results.csv'
with open(csv_filename) as f:
  reader = csv.DictReader(f)
  lst = list(reader)
  for item in lst:
    #item["query_results"] = ast.literal_eval(item["query_results"])
    if item["anterior_query_results"] != None:
      item["query_results"] = ast.literal_eval(item["query_results"])
      item["top_10_results"] = item["query_results"][:10]
      item["anterior_query_results"] = ast.literal_eval(item["anterior_query_results"])
    else:
      item["anterior_query_results"] = []
    movies.append(item)
```

```python
len(movies)
```

## Release

```python
best = 0
worst = 0
for movie in movies:
  if movie["no_prerelease"] == "1":
    if movie["success"] == "1":
      best += 1
    else:
      worst += 1
print("Best: {}, Worst: {}".format(best, worst))
```

# Movie Release Date

```python
releases = {}
for movie in movies:
  if movie["release_start"] not in releases.keys():
    releases[movie["release_start"]] = 0
  releases[movie["release_start"]] += 1  
releases = {k: v for k, v in sorted(releases.items(), key=lambda item: item[1])}
for release in list(releases.keys())[len(list(releases.keys()))-10:]:
    print("{} {}".format(release, releases[release]))
```

# Movie Distributors

```python
distributors = {}
for movie in movies:
  if movie["distributor"] not in distributors.keys():
    distributors[movie["distributor"]] = 0
  distributors[movie["distributor"]] += 1  
distributors = {k: v for k, v in sorted(distributors.items(), key=lambda item: item[1])}
for distributor in distributors:
    print("{} {}".format(distributor, distributors[distributor]))
```

# Movie Channels


```python
channels = {}
for movie in movies:
  for anterior_query_result in movie["top_10_results"]:
    if anterior_query_result[1] not in channels.keys():
      channels[anterior_query_result[1]] = 0
    channels[anterior_query_result[1]] += 1
channels = {k: v for k, v in sorted(channels.items(), key=lambda item: item[1])}
for channel in channels:
    print("{} {}".format(channel, channels[channel]))
```

```python
channels = {}
for movie in movies:
  for anterior_query_result in movie["top_10_results"]:
    if anterior_query_result[1] not in channels.keys():
      subs_str = anterior_query_result[6]
      if subs_str != None:
        if "K" in subs_str:
          subs_str = subs_str.replace("K", "")
          subs_str = float(subs_str) * 1000
        elif "M" in subs_str:
          subs_str = subs_str.replace("M", "")
          subs_str = float(subs_str) * 1000000
        else:
          subs_str = subs_str.replace("subscriber", "")
          subs_str = float(subs_str)
        
        subs_str = float(subs_str)

        channels[anterior_query_result[1]] = subs_str

channels = {k: v for k, v in sorted(channels.items(), key=lambda item: item[1])}

for channel in channels:
    print("{} {}".format(channel, channels[channel]))
```

```python
channels_verify = {}
for movie in movies:
    for anterior_query_result in movie["top_10_results"]:
        channel = anterior_query_result[1]
        if channel not in channels_verify.keys():
            if anterior_query_result[9] == "verified":
                channels_verify[channel] = True
            elif anterior_query_result[9] == "not verified":
                channels_verify[channel] = False
                #print("{} in {}".format(channel, anterior_query_result[9]))
            else:
                channels_verify[channel] = False
                #print("{} in {}".format(channel, anterior_query_result[9]))
```

```python
verified = 0
not_verified = 0
for channel in channels_verify:
    if channels_verify[channel]:
        verified += 1
    else:
        not_verified += 1

print("Verified {}, Not Verified {}".format(verified, not_verified))
```
